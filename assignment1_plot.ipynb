```python
# This is used to plot the average comparision of DSA Problem 1 in Assignment 1.

# This is the stadard library which is used for scientefic computation and deals with array.
import numpy as np
# this is the standard library which is used to plot graphs.
import matplotlib.pyplot as plt

txt_file = 'output_dsaq_1.txt'

# The below function is to read data from the text file which i have generated early and returns two lists, one list which stores array sizes and another one which stores average comparisons.
def rd_data(fname):
    # this list is to store the average number of compaision with respect ot the array size. 
    avg_comp = []
    # this list is to store the array size from the text file. 
    arra_sizes = []
    
    with open(fname, 'r') as reading_the_of_file:
        # this line of code help me to avoid the name of a coloum.
        next(reading_the_of_file)
        # this loop will iterate to each row of text file except the first row which will contain coloum name.
        for row in reading_the_of_file:
            # this will help me to slpit the rows.
            val = row.split()
            # the first part will assign into the one list know as arra_sizes.
            arra_sizes.append(int(val[0]))  
            # the second part will assign into the another list know as avg_comp.
            avg_comp.append(float(val[1]))

    # this will returns the two list
    return avg_comp, arra_sizes

# it will reads the data from the text file and makes saperate list for the comparision and size.
avg_comp, arra_sz = rd_data(txt_file)

# this is used to mentain equally space between bars on x-axis.
x_pos_spng = []
for i in range(len(arra_sz)):
    x_pos_spng.append(i) 

# this is defined the plotting area.
plt.figure(figsize=(11, 8))  
# it is help me to define the bar colour and width on the x-axis.
bars = plt.bar(x_pos_spng, avg_comp, color='blue', width=0.5)  

# Title for the bar graph
plt.title('Average Number of Comparison in Randomized Quick Sort Algorithm', fontsize=16)

# it is used to label the axis
plt.ylabel('Average Comparisons', fontsize=15)
plt.xlabel('Array Size', fontsize=14)

# Drawing the fine horizontal lines at each y-axis scales.
for i in plt.yticks()[0]:
    # it defines the line like thickness, colour, dashed or other things
    plt.axhline(y=i, color='black', linestyle='--', linewidth=0.6) 

# it annotes each bar with respect to its height.
for bar, size in zip(bars, arra_sz):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.05,  # Position just above the bar
             f'{int(bar.get_height())}', ha='center', va='bottom', fontsize=10, color='black')


# it is help me to customise x-ticks based on array size.
x_tick_labels = []
for size in arra_sz:
    x_tick_labels.append(str(size))
plt.xticks(x_pos_spng, x_tick_labels, fontsize=10)

# it helps to display the bar graph
plt.tight_layout()
plt.show()
```

```python
# This pyhton program is used to plot the average ratation and average height in the AVL and Red-Black tree.
# This is the Insertion in tree, DSA Problem 2 part 1 in Assignment 1.

# This is the stadard library which is used for scientefic computation and deals with array.
import numpy as np
# this is the standard library which is used to plot graphs.
import matplotlib.pyplot as plt

# this function is used to read data from the text file of red balck and avl tree, and it returns array size, avg_rotations, and avg_heights as lists for the tree int differnt text file.
def read_data_from_file(filename):
    array_sizes = []
    avg_rotations = []
    avg_heights = []
    with open(filename, 'r') as file:
        for row in file:
            data = list(filter(None, row.split()))
            
            # Skip the header or any non-numeric lines
            if data[0].isdigit():  # Only process if the first item is a number (tree size)
                size, rotation, height = data[:3]
                array_sizes.append(int(size))
                avg_rotations.append(float(rotation))
                avg_heights.append(float(height))
    return array_sizes, avg_rotations, avg_heights

# Reading data from the text files
rd_tree_txt_file = 'rb_tree_insertion.txt'
# calling the function which will extract the data from the text file and store it into the list.
rd_sizes, rd_avg_rotations, rd_avg_heights = read_data_from_file(rd_tree_txt_file)

# Reading data from the text files
avl_tree_txt_file = 'avl_tree_insertion.txt'
# calling the function which will extract the data from the text file and store it into the list.
avl_sizes, avl_avg_rotations, avl_avg_heights = read_data_from_file(avl_tree_txt_file)

# this is the function to add labels on the top of the each bars.
def add_bar_labels(rects):
    for rect in rects:
        height = rect.get_height()
        plt.text(rect.get_x() + rect.get_width() / 2., height,
                 f'{height:.2f}', ha='center', va='bottom')

# This fuction is used plot bar graph for average rotations, during insertion.
def plot_avg_rotations():
    # X positions for the bars
    index = np.arange(len(rd_sizes))

    # this defines the thickness of a bars in abar graph
    bar_width = 0.4  

    # Plot bars for Red-Black tree and AVL tree
    rd_bars = plt.bar(index, rd_avg_rotations, bar_width, label='AVL Tree', color='blue')
    avl_bars = plt.bar(index + bar_width, avl_avg_rotations, bar_width, label='Red-Black Tree', color='black')

    # using plt. adding an labels and titles into the bar graph.
    plt.title('Bar graph of average Rotations during inserting in Red-Black Tree & AVL Tree')
    plt.xlabel('Different Size of the array')
    plt.ylabel('Average Rotations values')

    # using this i am plot bars for both the Red-Black tree and AVL tree
    plt.xticks(index + bar_width / 2, [f'Size {s}' for s in rd_sizes], rotation=45)
    plt.legend()

    # Add value labels on top of each bar
    add_bar_labels(rd_bars)
    add_bar_labels(avl_bars)

    # shows the plot
    plt.tight_layout()

    # Drawing the fine horizontal lines at each y-axis scales.
    for i in plt.yticks()[0]:
        # it defines the line like thickness, colour, dashed or other things
        plt.axhline(y=i, color='black', linestyle='--', linewidth=0.6) 

    # using this to plot bar graph
    plt.show()

# this is the aperate function call for ploting bar graph for average rotation uisng both the trees.
plot_avg_rotations()

# Bar graph plotting for average heights (first and last columns: size, avg_height)
def plot_avg_heights():
    # X positions for the bars
    index = np.arange(len(rd_sizes))

    # this defines the thickness of a bars in abar graph
    bar_width = 0.4

    # using this i am plot bars for both the Red-Black tree and AVL tree
    rd_bars = plt.bar(index, rd_avg_heights, bar_width, label='Red-Black Tree', color='blue')
    avl_bars = plt.bar(index + bar_width, avl_avg_heights, bar_width, label='AVL Tree', color='black')

    # using plt. adding an labels and titles into the bar graph.
    plt.title('Bar graph of average Heights during inserting in Red-Black Tree & AVL Tree')
    plt.xlabel('Different Size of the array')
    plt.ylabel('Average Height Values')

    # Adding labels for each bar with tree name, size, and avg heights
    plt.xticks(index + bar_width / 2, [f'Size {s}' for s in rd_sizes], rotation=45)
    plt.legend()

    # Add value labels on top of each bar
    add_bar_labels(rd_bars)
    add_bar_labels(avl_bars)

    # shows the plot
    plt.tight_layout()

    # Drawing the fine horizontal lines at each y-axis scales.
    for i in plt.yticks()[0]:
        # it defines the line like thickness, colour, dashed or other things
        plt.axhline(y=i, color='black', linestyle='--', linewidth=0.6) 

    # using this to plot bar graph
    plt.show()

# this is the aperate function call for ploting bar graph for average height uisng both the trees.
plot_avg_heights()
```
